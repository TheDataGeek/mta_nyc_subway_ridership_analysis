IMPORT LIBRARIES

```python
import requests
import pandas as pd
import time
import os
```

HTTP GET REQUEST (LIMIT 1 ROW)

```python
response = requests.get(
    'https://data.ny.gov/resource/wujg-7c2s.json',
    params={'$limit': 1},
    timeout=300
)

for key, value in response.json()[0].items():
    print(f'{key}: {value}')
```

HTTP GET REQUEST 2021 MTA SUBWAY RIDERSHIP BY MONTH

```python
url = 'https://data.ny.gov/resource/wujg-7c2s.json?$limit=4000000'

months = [
    ('2021-01-01T00:00:00', '2021-01-31T23:59:59'),
    ('2021-02-01T00:00:00', '2021-02-28T23:59:59'),
    ('2021-03-01T00:00:00', '2021-03-31T23:59:59'),
    ('2021-04-01T00:00:00', '2021-04-30T23:59:59'),
    ('2021-05-01T00:00:00', '2021-05-31T23:59:59'),
    ('2021-06-01T00:00:00', '2021-06-30T23:59:59'),
    ('2021-07-01T00:00:00', '2021-07-31T23:59:59'),
    ('2021-08-01T00:00:00', '2021-08-31T23:59:59'),
    ('2021-09-01T00:00:00', '2021-09-30T23:59:59'),
    ('2021-10-01T00:00:00', '2021-10-31T23:59:59'),
    ('2021-11-01T00:00:00', '2021-11-30T23:59:59'),
    ('2021-12-01T00:00:00', '2021-12-31T23:59:59'),
]

all_months = []

for start, end in months:
    filename = f"../data/mta_{start[:7].replace('-', '_')}.csv"  

    if os.path.exists(filename):
        print(f"Skipping {filename} — already exists")
        all_months.append(pd.read_csv(filename))
        continue

    params = {
        '$where': f"transit_mode='subway' AND transit_timestamp >= '{start}' AND transit_timestamp <= '{end}'"
    }
    response = requests.get(url, params=params, timeout=300)
    response.raise_for_status()
    
    month_df = pd.DataFrame(response.json())
    month_df.to_csv(filename, index=False)
    
    print(f"Saved {filename} — {len(month_df):,} rows")
    
    all_months.append(month_df)
    time.sleep(5)

print('Data collection complete!')
```
