IMPORT LIBRARIES

```python
import pandas as pd
```

LOAD/COMBINE DATASETS

```python
jan = pd.read_csv('../data/mta_2021_01.csv')
feb = pd.read_csv('../data/mta_2021_02.csv')
mar = pd.read_csv('../data/mta_2021_03.csv')
apr = pd.read_csv('../data/mta_2021_04.csv')
may = pd.read_csv('../data/mta_2021_05.csv')
jun = pd.read_csv('../data/mta_2021_06.csv')
jul = pd.read_csv('../data/mta_2021_07.csv')
aug = pd.read_csv('../data/mta_2021_08.csv')
sep = pd.read_csv('../data/mta_2021_09.csv')
oct = pd.read_csv('../data/mta_2021_10.csv')
nov = pd.read_csv('../data/mta_2021_11.csv')
dec = pd.read_csv('../data/mta_2021_12.csv')

# CONCATENATE DATASETS
df = pd.concat([jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec], ignore_index=True)

print(f"Total rows: {len(df):,}")
```

```python
df.info()
```

DATE STANDARDIZATION AND COLUMN ENGINEERING

```python
# CHECK 'TRANSIT_TIMESTAMP' FORMAT
print(df['transit_timestamp'][0])
print(type(df['transit_timestamp'][0]))
```

```python
# CONVERT TO DATETIME
df['transit_timestamp'] = pd.to_datetime(df['transit_timestamp'], format='ISO8601')

# DATETIME COLUMN ENGINEERING
df['year']        = df['transit_timestamp'].dt.year
df['month']       = df['transit_timestamp'].dt.month
df['month_name']  = df['transit_timestamp'].dt.strftime('%B')
df['day']         = df['transit_timestamp'].dt.day
df['hour']        = df['transit_timestamp'].dt.hour
df['day_of_week'] = df['transit_timestamp'].dt.day_name()
df['date']        = df['transit_timestamp'].dt.date
```

```python
df.info()
```

```python
# CHECK 'TRANSIT_MODE', 'STATION_COMPLEX_ID'
print(df[['transit_mode','station_complex_id']].iloc[0])
```

```python
# CHECK 'LATITUDE', 'LONGITUDE' FORMAT
print(df[['latitude','longitude']].iloc[0])
```

```python
# CHECK 'GEOREFERENCE' FORMAT
print(df['georeference'].iloc[0])
print(type(df['georeference'].iloc[0]))
```

```python
# DROP UNNECESSARY COLUMNS
df.drop(columns = ['transit_mode','station_complex_id','georeference'], inplace = True)
```

```python
df.info()
```

```python
# RETURN COUNT OF UNIQUE VALUES
df['station_complex'].nunique()
```

```python
# RETURN SORTED UNIQUE STATIONS
unique_stations = sorted(df['station_complex'].unique())
unique_stations
```

```python
# COMBINE 34 ST-PENN STATION ENTRANCES
df['station_complex'] = df['station_complex'].replace({
    '34 St-Penn Station (A,C,E)': '34 St-Penn Station',
    '34 St-Penn Station (1,2,3)': '34 St-Penn Station'
})
```

```python
# VERIFY COUNT OF UNIQUE VALUES (SHOULD BE LESS 1)
df['station_complex'].nunique()
```

```python
df.info()
```

EXPORT CLEANED TO CSV

```python
df.to_csv('../data/mta_2021_clean.csv', index=False)

print(f"Total rows: {len(df):,}")
print('Saved: ../data/mta_2021_clean.csv')
```

BASIC EDA AND STATS

```python
# DESCRIPTIVE STATS
df_clean.describe()
```

```python
# BASIC RIDERSHIP ANALYSIS
total_ridership = df_clean['ridership'].sum()
total_transfers = df_clean['transfers'].sum()
total_stations  = df_clean['station_complex'].nunique()

omny_ridership  = df_clean[df_clean['payment_method'] == 'omny']['ridership'].sum()
metro_ridership = df_clean[df_clean['payment_method'] == 'metrocard']['ridership'].sum()

omny_pct  = omny_ridership / total_ridership * 100
metro_pct = metro_ridership / total_ridership * 100

print('--- Ridership by the Numbers ---')
print(f"Total Ridership:     {total_ridership:,.0f}")
print(f"Total Transfers:     {total_transfers:,.0f}")
print(f"Total Stations:      {total_stations}")
print(f"OMNY Ridership:      {omny_ridership:,.0f}")
print(f"MetroCard Ridership: {metro_ridership:,.0f}")
print(f"OMNY %:              {omny_pct:.1f}%")
print(f"MetroCard %:         {metro_pct:.1f}%")
```

```python
# BY BOROUGH
print('--- Ridership by Borough ---')
borough = (
    df_clean.groupby('borough')['ridership']
    .sum()
    .sort_values(ascending=False)
)
print(borough)
```

```python
# BY MONTH
print('--- Ridership by Month ---')
monthly = (
    df_clean.groupby(['month', 'month_name'])['ridership']
    .sum()
    .reset_index()
    .sort_values('month')
)
print(monthly)
```

```python
# BY WEEK
print('--- Ridership by Day of Week ---')
daily = (
    df_clean.groupby('day_of_week')['ridership']
    .sum()
    .reindex(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
)
print(daily)
```

```python
# BY HOUR
print('--- Ridership by Hour (5AM - 12AM) ---')
hourly_filtered = (
    df_clean.groupby('hour')['ridership']
    .sum()
    .reindex(range(5, 24))  # 5AM to 11PM (23:00)
    .reset_index()
)
print(hourly_filtered)
```

```python
# BUSIEST STATION, MONTH, WEEK, HOUR
print('--- Busiest Station ---')
busiest_station = df_clean.groupby('station_complex')['ridership'].sum().idxmax()
busiest_station_total = df_clean.groupby('station_complex')['ridership'].sum().max()
print(f"{busiest_station}: {busiest_station_total:,.0f}")

print()
print('--- Busiest Month ---')
busiest_month = df_clean.groupby('month_name')['ridership'].sum().idxmax()
busiest_month_total = df_clean.groupby('month_name')['ridership'].sum().max()
print(f"{busiest_month}: {busiest_month_total:,.0f}")

print()
print('--- Busiest Day of Week ---')
busiest_day = df_clean.groupby('day_of_week')['ridership'].sum().idxmax()
busiest_day_total = df_clean.groupby('day_of_week')['ridership'].sum().max()
print(f"{busiest_day}: {busiest_day_total:,.0f}")

print()
print('--- Busiest Hour ---')
busiest_hour = df_clean.groupby('hour')['ridership'].sum().idxmax()
busiest_hour_total = df_clean.groupby('hour')['ridership'].sum().max()
print(f"{busiest_hour}: {busiest_hour_total:,.0f}")
```

```python
# TOP 10 BUSIEST STATIONS
print('--- Top 10 Busiest Stations ---')
top_stations = (
    df_clean.groupby(['station_complex', 'borough'])['ridership']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

print(top_stations)
```

```python
# CLEAN STATION NAMES
top_stations['station_name'] = (
    top_stations['station_complex']
    .str.replace(r'\s*\(.*?\)', '', regex=True)
    .str.strip()
)

print(top_stations[['station_name', 'borough', 'ridership']])
```

```python
# BY FARE
print('--- Fare Class Breakdown ---')
fare = (
    df_clean.groupby('fare_class_category')['ridership']
    .sum()
    .sort_values(ascending=False)
)
print(fare)
```
